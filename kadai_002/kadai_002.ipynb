```python
import requests
from getpass import getpass

# 1. APIキーと駅名の入力を受け付ける
api_key = getpass('APIキーを入力してください: ')
search_station = input('検索する駅を入力してください: ')

# --- ステップ1: 駅の緯度経度を取得 ---
geocode_url = "https://maps.googleapis.com/maps/api/geocode/json"
geocode_params = {
    "address": search_station,
    "key": api_key
}

geocode_response = requests.get(geocode_url, params=geocode_params).json()

if geocode_response["status"] == "OK":
    location = geocode_response["results"][0]["geometry"]["location"]
    lat = location["lat"]
    lng = location["lng"]
    print(f"\n{search_station}駅の座標: 緯度 {lat}, 経度 {lng}")

    # --- ステップ2: 半径500m以内のレストラン情報を取得 ---
    place_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    place_params = {
        "location": f"{lat},{lng}",
        "radius": 500,
        "type": "restaurant",
        "language": "ja",
        "key": api_key
    }

    place_response = requests.get(place_url, params=place_params).json()

    # --- ステップ3: 結果を出力 ---
    if place_response["status"] == "OK":
        print("\n--- 周辺のレストラン情報 ---")
        for place in place_response["results"]:
            name = place.get("name", "名称不明")
            rating = place.get("rating", "評価なし")
            address = place.get("vicinity", "住所不明")

            print(f"店名: {name}")
            print(f"評価: {rating}")
            print(f"住所: {address}")
            print("-" * 20)
    else:
        print("レストラン情報が見つかりませんでした。")
else:
    print(f"駅が見つかりませんでした。エラー: {geocode_response.get('status')}")
```
