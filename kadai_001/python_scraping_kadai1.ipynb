```python
import requests
from bs4 import BeautifulSoup

# 記事の本文を取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/3495e3a75306f03f4b786b2404a3712baba2aaf5'

#GETリクエストを送信し、HTMLを取得する
response = requests.get(url)

#HTMLを解析する
soup = BeautifulSoup(response.text,'html.parser')

#記事の本文を取得する
#CSSセレクタを使用して、記事本文を取得する
body_paragraphs = soup.select('p.sc-54nboa-0.zUIkh.yjSlinkDirectlink.highLightSearchTarget')

#記事の本文テキストを取得する
# 要素が見つからなかった場合のためにNoneチェックを追加
if body_paragraphs:
    # 取得したすべての段落のテキストを結合する
    body_text = '\n'.join([p.get_text(strip=True) for p in body_paragraphs])
else:
    body_text = "記事の本文が見つかりませんでした。"

#テキストを表示する
print(body_text)
```
