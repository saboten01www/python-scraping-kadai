```python
import requests
from bs4 import BeautifulSoup

# 記事の本文を取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/3495e3a75306f03f4b786b2404a3712baba2aaf5'

#GETリクエストを送信し、HTMLを取得する
response = requests.get(url)

#HTMLを解析する
soup = BeautifulSoup(response.text,'html.parser')

# 不要な段落を削除する（記事本文の取得前に実行）
# これらの要素は記事本文とは別の場所にあるか、誤って本文として選択される可能性のあるものと仮定
for unwanted in soup.select('p.sc-uzx6gd-4.kxaOcP'): # 不要なクラス名を指定
    unwanted.decompose()
for unwanted in soup.select('p.sc-wwbbh8-0.branMF'): # 不要なクラス名を指定
    unwanted.decompose()

#記事の本文を取得する
#CSSセレクタを使用して、記事本文を取得する
body_paragraphs = soup.select('p.sc-54nboa-0.zUIkh.yjSlinkDirectlink.highLightSearchTarget')

if body_paragraphs:
    # 2. 本文内にある不要な<a>タグをすべて探して削除する
    # 例：特定のクラスを持つリンクや、特定の文字列を含むリンクを削除
    for paragraph in body_paragraphs: # Iterate through each paragraph
        for a_tag in paragraph.find_all('a'): # Call find_all on the individual paragraph
            # 必要に応じて条件を絞り込む
            if '写真で見る' in a_tag.get_text():
                a_tag.decompose()
            if '解説' in a_tag.get_text():
                a_tag.decompose()



#記事の本文テキストを取得する
# 要素が見つからなかった場合のためにNoneチェックを追加
if body_paragraphs:
    # 取得したすべての段落のテキストを結合する
    body_text = '\n'.join([p.get_text(strip=True) for p in body_paragraphs])
else:
    body_text = "記事の本文が見つかりませんでした。"

#テキストを表示する
print(body_text)
```
